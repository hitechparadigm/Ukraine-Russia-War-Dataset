```python
import pandas as pd

# Load the datasets
weather_data = pd.read_csv('../data/daily_weather.csv')
casualties_data = pd.read_csv('../data/df_daily_casualties_russia.csv')

# Display the first few rows of each dataframe to understand their structure
weather_data_head = weather_data.head()
casualties_data_head = casualties_data.head()

weather_data_head, casualties_data_head
```

```python
# Convert the date formats to be consistent
weather_data['day'] = pd.to_datetime(weather_data['day'])
casualties_data['date'] = pd.to_datetime(casualties_data['date'], format='%d.%m.%Y')

# Merge the datasets on the date columns
merged_data = pd.merge(weather_data, casualties_data, left_on='day', right_on='date', how='inner')

# Drop the redundant date column
merged_data = merged_data.drop(columns=['date'])

merged_data.tail()
```

```python
merged_data.to_csv('../data/merged_data_casualties_weather.csv', index=False)
```
